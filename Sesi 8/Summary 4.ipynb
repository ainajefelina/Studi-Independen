# DATA INSIDEN DI DEPARTEMEN POLISI SAN FRANCISCO

- mengimpor data yang dibutuhkan

```python
incidents = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv')
incidents.head()
```

```python
incidents.shape
```

- kita tahu pada tahun 2016 terjadi 150500 insiden
- kita dapat melihat 100 insiden pertama saja

```python
limit = 100
df_100incidents= incidents.iloc[0:limit, :]
```

```python
df_100incidents.shape
```

- karena kita akan memetakan titik terjadinya insiden maka perlu dilakukan pengurangan data

```python
latitude = 37.77
longitude = -122.42
```

```python
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)
sanfran_map
```

- dapat meletakan lokasi kejahatan dengan menambahkan dataframe

```python
incidents=folium.map.FeatureGroup()

for lat, lng, in zip(df_100incidents.Y, df_100incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng], 
            radius=5, 
            color='yellow', 
            fill=True, 
            fill_color='blue', 
            fill_opacity=0.6
        )
    )

sanfran_map.add_child(incidents)
```

```python
incidents=folium.map.FeatureGroup()

for lat, lng, in zip(df_100incidents.Y, df_100incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng], 
            radius=5, 
            color='yellow', 
            fill=True, 
            fill_color='blue', 
            fill_opacity=0.6
        )
    )

latitudes = list(df_100incidents.Y)
longitudes = list(df_100incidents.X)
labels = list(df_100incidents.Category)

for lat, lng, label, in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)

sanfran_map.add_child(incidents)
```

- marker terlalu berkerumun kita adalah dengan menghapus marker
- atau mengelompokkannya ke dalam cluster yang berbeda dengan mewakili jumlah kejahatan di setiap lingkungan

```python
from folium import plugins
```

```python
sanfran_map = folium.Map(location = [latitude, longitude], zoom_start = 12)

incidents1 = plugins.MarkerCluster().add_to(sanfran_map)

for lat, lng, label, in zip(df_100incidents.Y, df_100incidents.X, df_100incidents.Category): 
    folium.Marker(location=[lat, lng], icon=None, popup=label,).add_to(incidents1)

sanfran_map
```
