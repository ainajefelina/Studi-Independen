###### Penambahan data tahun, bulan, dan hari dari DatetimeIndex yang ada

```python
opsd_daily['Year'] = opsd_daily.index.year
opsd_daily['Month'] = opsd_daily.index.month
opsd_daily['Weekday'] = opsd_daily.index.weekday
```

###### dapat memvisualisasikan data time series yang berguna untuk plot time series

###### Menunjukan konsumsi listrik jerman

```python
opsd_daily['Consumption'].plot(linewidth=-.5)
```

```python
opsd_daily['Consumption'].plot(marker='.', alpha = 0.5, linestyle = 'None', figsize = (11, 9));
```

```python
opsd_daily['Solar'].plot(marker= '.', alpha = 0.5, linestyle = 'None', figsize = (11,9));
```

###### terdapat beberapa data yang menarik
- konsumsi listrik tertinggi di musim dingin dengan kemungkinan karena pemanas listrik dan peningkatan penggunaan penerangan, dan paling rendah di musim panas
- produksi tenaga surya paling tinggi di musim panas dikarenakan sinar matahari yang paling melimpah, dan rendah di musim dingin
- terjadi tren peningkatan dalam produksi tenaga angin selama bertahun - tahun

###### Periodisitas
- terjadi pola berulang pada interval waktu yang teratur
- Consumption, Solar, dan Wind berosilasi antara nilai tinggi dan rendah dalam skala waktu tahunan sesuai dengan berubahan cuaca musiman

```python
ax = opsd_daily.loc['2017', 'Consumption'].plot(figsize=(11,9))
ax.set_ylabel('Daily Consumption (GWh)');
```

```python
axx = opsd_daily.loc['2017-01':'2017-02', 'Consumption'].plot(marker='o', linestyle= '-', figsize=(11,9))
axx.set_ylabel('Daily Consumption (GWh)');
```

```python
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
opsd_weekly_mean = opsd_daily[data_columns].resample('W').mean()
```
